```python
import json
import time
import random
import os

```

```python
input_path = "/tmp/input_json"

names = ["Alice", "Bob", "Charlie", "Diana"]
cities = ["New York", "London", "Paris", "Berlin"]

os.makedirs("/tmp/input_json", exist_ok=True)
```

```python
for i in range(1, 6):
    # Create dummy data
    data = {
        "name": random.choice(names),
        "age": random.randint(20, 50),
        "city": random.choice(cities),
        "timestamp": time.time()
    }
    
    # Create a unique file name
    file_name = f"data_{random.randint(1000, 9999)}.json"
    full_path = os.path.join(input_path, file_name)
    
    # Write the file
    with open(full_path, 'w') as f:
        json.dump(data, f)
        
    print(f"Created file: {file_name}")
    time.sleep(2) # Wait 2 seconds before creating the next oneG
```
