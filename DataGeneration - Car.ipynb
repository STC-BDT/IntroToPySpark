```python
import json
import time
import random
import os
```

```python
output_dir = "/tmp/car_speeds"
car_ids = ["CAR001", "CAR002", "CAR003", "CAR004", "CAR005"]
speed_min = 30
speed_max = 120
lines_per_file = 5

os.makedirs(output_dir, exist_ok=True)
```

```python
line_counter = 0
current_file = None

while True:
    # Switch to a new file every 5 lines
    if line_counter % lines_per_file == 0:
        # Close previous file if it exists
        if current_file is not None:
            current_file.close()
        
        # Create new file with timestamp in name
        timestamp = int(time.time())
        filename = f"car_speeds_{timestamp}.jsonl"
        filepath = os.path.join(output_dir, filename)
        current_file = open(filepath, 'w')
        print(f"Created new file: {filename}")
    
    # Generate car data
    data = {
        "car_id": random.choice(car_ids),
        "car_speed": random.randint(speed_min, speed_max),
        "timestamp": time.time()
    }
    
    # Write JSON line to file
    current_file.write(json.dumps(data) + "\n")
    current_file.flush()  # Ensure data is written immediately
    
    line_counter += 1
    print(f"Written line {line_counter} to {os.path.basename(current_file.name)}")
    
    time.sleep(0.5)  # Wait 2 seconds before creating the next record
```
